```python
import numpy as np
import pandas as pd
import sklearn
```

```python
df = pd.read_csv("../data/abalone.data", delimiter=" ", header=None)
df
```

```python
from sklearn.neighbors import KNeighborsRegressor
X = df.iloc[:, 1:8]
y = df.iloc[:,[8]]
y = y.astype('float64')
neigh = KNeighborsRegressor(n_neighbors=20, algorithm='brute')
neigh.fit(X,y)
test_1 = [[0.71, 0.555,0.195,1.9485,0.9455,0.3765,0.4]]
print(neigh.predict(test_1))
l = neigh.kneighbors(test_1, n_neighbors=20, return_distance=False)
df.iloc[l[0]][8].sum() / 20
```

```python
from sklearn.cluster import KMeans
X = df.iloc[:, 1:]
kmeans = KMeans(n_clusters=5, init=[
    [0.455,0.365,0.095,0.514,0.2245,0.101,0.15,15],
[0.45,0.35,0.13,0.547,0.245,0.1405,0.1405,8],
[0.61,0.485,0.15,1.2405,0.6025,0.2915,0.3085,12],
[0.34,0.255,0.075,0.18,0.0745,0.04,0.0525,6],
[0.54,0.435,0.145,0.97,0.4285,0.22,0.264,17]
], max_iter=1).fit(X)
kmeans.cluster_centers_
```
